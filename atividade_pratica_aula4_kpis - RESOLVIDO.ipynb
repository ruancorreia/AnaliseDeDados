## 4. KPIs globais da operação

Com base nas “4 Métricas de Ouro do Varejo” apresentadas na aula, calcule os seguintes KPIs globais da base inteira:

- **Receita Total**
- **Lucro Total**
- **Margem de Lucro (%)**
- **Ticket Médio**

### Orientação
- Pense primeiro na fórmula de cada KPI
- Escreva os cálculos em células separadas ou em uma pequena tabela-resumo
- Se quiser, formate os resultados para facilitar a leitura


```python
import pandas as pd
from pathlib import Path

csv_path = Path('vendas_brasil_aula4_kpis.csv')

if 'df' not in globals():
    df = pd.read_csv(csv_path)

if 'data_venda' in df.columns:
    df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# Calcule aqui os 4 KPIs globais
receita_total = df['receita'].sum()
lucro_total = df['lucro'].sum()
margem_lucro = lucro_total / receita_total if receita_total else 0
ticket_medio = receita_total / df['pedido_id'].nunique() if 'pedido_id' in df.columns else df['receita'].mean()

kpis_globais = pd.DataFrame({
    'KPI': ['Receita Total', 'Lucro Total', 'Margem de Lucro', 'Ticket Médio'],
    'Valor': [receita_total, lucro_total, margem_lucro, ticket_medio]
})

kpis_globais['Valor'] = [
    f'R$ {receita_total:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
    f'R$ {lucro_total:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
    f'{margem_lucro:.2%}'.replace('.', ','),
    f'R$ {ticket_medio:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')
]

kpis_globais
```

### Interpretação
O negócio parece de porte relevante, com receita total acima de R$ 1,9 milhão e lucro total superior a R$ 544 mil.
A margem de lucro em torno de 28% indica uma operação saudável, com boa conversão da receita em resultado.
O ticket médio próximo de R$ 4,6 mil sugere vendas de valor alto por pedido, o que reforça um perfil de operação mais consultiva e menos pulverizada.
Em resumo, a receita está sendo bem convertida em lucro, mas vale acompanhar a eficiência por canal e por categoria para sustentar esse desempenho.


## 5. Métricas derivadas

A aula mostra que **somar valores não basta**.  
Precisamos criar métricas derivadas para avaliar eficiência e comportamento.

### Tarefa
Crie, quando fizer sentido:
- uma coluna de **margem_lucro** (`lucro / receita`)
- uma lógica para **ticket_medio** da operação ou por grupo

### Reflexão
Explique:
1. Por que receita alta não significa automaticamente melhor desempenho?
2. Em que situação a margem ajuda mais do que a receita bruta?


```python
# Crie aqui as métricas derivadas
df['margem_lucro'] = df['lucro'] / df['receita']
ticket_medio_operacao = df['receita'].sum() / df['pedido_id'].nunique() if 'pedido_id' in df.columns else df['receita'].mean()

ticket_medio_por_canal = (
    df.groupby('canal_venda')['receita'].sum() / df.groupby('canal_venda')['pedido_id'].nunique()
) if 'pedido_id' in df.columns else df.groupby('canal_venda')['receita'].mean()

metricas_derivadas = pd.DataFrame({
    'margem_lucro_media': [df['margem_lucro'].mean()],
    'ticket_medio_operacao': [ticket_medio_operacao]
})

metricas_derivadas

ticket_medio_por_canal
```

### Reflexão respondida
1. Receita alta não significa automaticamente melhor desempenho porque uma operação pode vender muito e ainda assim ter custos elevados, reduzindo a eficiência do resultado final.
2. A margem ajuda mais do que a receita bruta quando é preciso comparar eficiência entre canais, categorias ou produtos, porque mostra quanto realmente sobra de cada venda.

## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
import pandas as pd
from pathlib import Path

if 'df' not in globals():
    csv_path = Path('vendas_brasil_aula4_kpis.csv')
    df = pd.read_csv(csv_path)

tabela_canal = (
    df.groupby('canal_venda')
      .agg(Receita_Total=('receita', 'sum'),
           Lucro_Total=('lucro', 'sum'),
           Pedidos=('canal_venda', 'size'))
      .assign(Margem=lambda x: x['Lucro_Total'] / x['Receita_Total'],
              Ticket_Medio=lambda x: x['Receita_Total'] / x['Pedidos'])
      .reset_index()
      .sort_values('Receita_Total', ascending=False)
)

tabela_canal['Receita_Total'] = tabela_canal['Receita_Total'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
tabela_canal['Lucro_Total'] = tabela_canal['Lucro_Total'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
tabela_canal['Margem'] = tabela_canal['Margem'].map(lambda v: f'{v:.2%}'.replace('.', ','))
tabela_canal['Ticket_Medio'] = tabela_canal['Ticket_Medio'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

tabela_canal
```

### Insight obrigatório
O Marketplace concentra a maior receita, mas o canal Online apresenta a melhor margem e o maior ticket médio, indicando maior eficiência por venda.
Na prática, a decisão não deve priorizar apenas volume: faz sentido proteger o canal de maior faturamento e, ao mesmo tempo, ampliar ações no canal Online, que entrega melhor rentabilidade por pedido.

## 10. O motor da análise: agregações (groupby)

Nos slides, a lógica central é:
- `receita -> sum`
- `margem -> mean` ou cálculo derivado
- `produto -> count`

### Tarefa
Escolha **duas dimensões diferentes** e construa tabelas agregadas mostrando que você entendeu essa lógica.

Exemplos:
- por canal
- por categoria
- por UF
- por mês


```python
if 'margem_lucro' not in df.columns:
    df['margem_lucro'] = df['lucro'] / df['receita']

agregacao_canal = (
    df.groupby('canal_venda')
      .agg(Receita_Total=('receita', 'sum'),
           Margem_Media=('margem_lucro', 'mean'),
           Qtd_Produtos=('produto', 'count'))
      .sort_values('Receita_Total', ascending=False)
)

df['mes'] = pd.to_datetime(df['data_venda'], errors='coerce').dt.to_period('M').astype(str)
agregacao_mes = (
    df.groupby('mes')
      .agg(Receita_Total=('receita', 'sum'),
           Margem_Media=('margem_lucro', 'mean'),
           Qtd_Produtos=('produto', 'count'))
      .sort_index()
)

agregacao_canal_view = agregacao_canal.copy()
agregacao_canal_view['Receita_Total'] = agregacao_canal_view['Receita_Total'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
agregacao_canal_view['Margem_Media'] = agregacao_canal_view['Margem_Media'].map(lambda v: f'{v:.2%}'.replace('.', ','))

agregacao_mes_view = agregacao_mes.copy()
agregacao_mes_view['Receita_Total'] = agregacao_mes_view['Receita_Total'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
agregacao_mes_view['Margem_Media'] = agregacao_mes_view['Margem_Media'].map(lambda v: f'{v:.2%}'.replace('.', ','))

agregacao_canal_view, agregacao_mes_view
```

## 11. Campeões e vilões (Top N)

A aula destaca o poder dos rankings curtos e focados.

### Tarefa
Crie:
- Top 10 produtos por **lucro**
- Top 10 produtos por **receita**
- 5 categorias ou UFs com **pior margem** ou pior desempenho

### Perguntas
1. Quem são os “campeões”?
2. Quem são os “vilões”?
3. Que ação tática poderia ser tomada com base nisso?

### Respostas
Os campeões são Notebook Pro, Smartphone X e Tablet Plus, que lideram tanto em lucro quanto em receita.
Como vilões de eficiência, as categorias de maior receita (Telefonia e Informática) têm margens menores que Áudio e Periféricos.
Ação tática: preservar volume nos campeões e revisar preço, mix e custo nas categorias de menor margem para elevar rentabilidade sem perder faturamento.


```python
top10_lucro = (
    df.groupby('produto', as_index=False)['lucro']
      .sum()
      .sort_values('lucro', ascending=False)
      .head(10)
)

top10_receita = (
    df.groupby('produto', as_index=False)['receita']
      .sum()
      .sort_values('receita', ascending=False)
      .head(10)
)

piores_categorias_margem = (
    df.groupby('categoria', as_index=False)
      .agg(receita=('receita', 'sum'), lucro=('lucro', 'sum'))
      .assign(margem=lambda x: x['lucro'] / x['receita'])
      .sort_values('margem', ascending=True)
      .head(5)
)

top10_lucro_view = top10_lucro.copy()
top10_lucro_view['lucro'] = top10_lucro_view['lucro'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

top10_receita_view = top10_receita.copy()
top10_receita_view['receita'] = top10_receita_view['receita'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

piores_categorias_margem_view = piores_categorias_margem.copy()
piores_categorias_margem_view['receita'] = piores_categorias_margem_view['receita'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
piores_categorias_margem_view['lucro'] = piores_categorias_margem_view['lucro'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
piores_categorias_margem_view['margem'] = piores_categorias_margem_view['margem'].map(lambda v: f'{v:.2%}'.replace('.', ','))

top10_lucro_view, top10_receita_view, piores_categorias_margem_view
```

## 12. Análise temporal

A aula reforça que sem tempo a análise fica estática.

### Tarefa
1. Converta `data_venda` para data, se necessário
2. Crie colunas auxiliares como:
   - mês
   - trimestre (opcional)
3. Gere uma tabela agregada por mês com:
   - receita
   - lucro
   - ticket médio ou margem

### Depois responda
1. O lucro parece crescer ao longo do tempo?
2. O último trimestre foi melhor que o anterior?
3. Quais meses concentram as vendas de fim de ano?

### Respostas
O lucro não cresce de forma linear, mas há aceleração no segundo semestre com picos em novembro e dezembro.
O 4º trimestre supera o 3º trimestre em receita e lucro, sinalizando sazonalidade positiva de fim de ano.
Os meses que mais concentram vendas de fim de ano são novembro e dezembro.


```python
if 'margem_lucro' not in df.columns:
    df['margem_lucro'] = df['lucro'] / df['receita']

df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')
df['mes'] = df['data_venda'].dt.to_period('M').astype(str)
df['trimestre'] = df['data_venda'].dt.to_period('Q').astype(str)

analise_mensal = (
    df.groupby('mes')
      .agg(receita=('receita', 'sum'),
           lucro=('lucro', 'sum'),
           margem=('margem_lucro', 'mean'))
      .reset_index()
      .sort_values('mes')
)

analise_trimestral = (
    df.groupby('trimestre')
      .agg(receita=('receita', 'sum'), lucro=('lucro', 'sum'))
      .reset_index()
      .sort_values('trimestre')
)

analise_mensal_view = analise_mensal.copy()
analise_mensal_view['receita'] = analise_mensal_view['receita'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
analise_mensal_view['lucro'] = analise_mensal_view['lucro'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
analise_mensal_view['margem'] = analise_mensal_view['margem'].map(lambda v: f'{v:.2%}'.replace('.', ','))

analise_trimestral_view = analise_trimestral.copy()
analise_trimestral_view['receita'] = analise_trimestral_view['receita'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
analise_trimestral_view['lucro'] = analise_trimestral_view['lucro'].map(lambda v: f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))

analise_mensal_view, analise_trimestral_view
```

```python
resumo_executivo = pd.DataFrame({
    'KPI': ['Receita Total', 'Lucro Total', 'Margem', 'Ticket Médio'],
    'Valor': [
        df['receita'].sum(),
        df['lucro'].sum(),
        df['lucro'].sum() / df['receita'].sum(),
        df['receita'].sum() / df['pedido_id'].nunique()
    ]
})

resumo_canal = (
    df.groupby('canal_venda')
      .agg(Receita_Total=('receita', 'sum'),
           Lucro_Total=('lucro', 'sum'),
           Pedidos=('pedido_id', 'nunique'))
      .assign(Margem=lambda x: x['Lucro_Total'] / x['Receita_Total'],
              Ticket_Medio=lambda x: x['Receita_Total'] / x['Pedidos'])
      .reset_index()
)

def formata_moeda(v):
    return f'R$ {v:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')

resumo_executivo_view = resumo_executivo.copy()
resumo_executivo_view['Valor'] = resumo_executivo_view.apply(
    lambda row: f"{row['Valor']:.2%}".replace('.', ',') if row['KPI'] == 'Margem' else formata_moeda(row['Valor']),
    axis=1
)

resumo_canal_view = resumo_canal.copy()
resumo_canal_view['Receita_Total'] = resumo_canal_view['Receita_Total'].map(formata_moeda)
resumo_canal_view['Lucro_Total'] = resumo_canal_view['Lucro_Total'].map(formata_moeda)
resumo_canal_view['Margem'] = resumo_canal_view['Margem'].map(lambda v: f'{v:.2%}'.replace('.', ','))
resumo_canal_view['Ticket_Medio'] = resumo_canal_view['Ticket_Medio'].map(formata_moeda)

resumo_executivo_view, resumo_canal_view
```
